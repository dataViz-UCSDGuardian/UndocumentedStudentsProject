```python
import matplotlib.pyplot as plt 
import numpy as np

def pie(ax, values, **kwargs):
    total = sum(values)
    def formatter(pct):
        return '{:0.0f}'.format(pct*total/100, pct)
    wedges, _, labels = ax.pie(values, autopct=formatter,**kwargs)
    return wedges

fig, ax = plt.subplots()
ax.axis('equal')

width = 0.5 
kwargs = dict(colors=['#1982B1', '#D6CE78', '#F4A05A'], startangle=90, counterclock=False)

outside = pie(ax, [3044,21372,55772], radius=2.1, pctdistance=1-width/4, **kwargs)
inside = pie(ax, [400,3600,34000], radius=2-width, 
             pctdistance=1 - (width/4) / (1-width/2), **kwargs)
plt.setp(inside + outside, width=width, edgecolor='white')

ax.legend(['University of California', 'California State University', 'Community College'],bbox_to_anchor=(1.75,1.5),frameon=False)

kwargs = dict(size=20, color='black', va='center', fontweight='bold')
ax.text(0, 3, 'Distribution of Undocumented Students \n Enrolled in California Colleges', ha='center', **kwargs)

descript = 'Figure 1: Actual enrollment of Undocumented Students\namong California Colleges before and after\nthe Dream Act was passed on December 8, 2010'
ax.text(0,-3,descript, va='bottom', ha='center')

an1 = ax.annotate("Before Dream Act", xy=(0, 0),
                  va="center", ha="center", size=14,
                  bbox=dict(boxstyle="round", fc="w"))

an2 = ax.annotate("After Dream Act", xy=(1.05,0.5), xycoords=an1,
                  xytext=(2.2,0),
                  va="bottom", ha="left", size=14,
                  bbox=dict(boxstyle="round", fc="w"),
                  arrowprops=dict(arrowstyle="<-"))
plt.show()
```
